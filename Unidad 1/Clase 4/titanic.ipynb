```python
hombres_sobrevivieron = df[(df["Survived"] == 1) & (df["Sex"] == "male")]

# mujeres_sobrevivieron.head()
print(len(hombres_sobrevivieron))
```

```python
jovenes = df[(df["Age"] >= 0) & (df["Age"] <= 17)]

print(len(jovenes))
```

```python
adultos = df[(df["Age"] >= 18) & (df["Age"] <= 50)]

print(len(adultos))
```

```python
adultos_mayor = df[df["Age"] > 50]

print(len(adultos_mayor))
```

```python
import matplotlib.pyplot as plt
```

```python
# contenido 
total_adultos = len(adultos)
total_adultos_supervivientes = len(AdultosSuperviventes)

#Datos para la grafica
etiquetas =  ['Adultos', 'Adultos Supervivientes']
valores = [total_adultos,total_adultos_supervivientes]

#Creagar Grafica 
plt.figure(figsize=(6,4))
plt.bar(etiquetas,valores, color = ['gray','green'])
plt.title("Compracion: Adultos Totales vs Adultos Supervivientes")
plt.ylabel("Numero de personas")
plt.xlabel("Categorias")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# contenido 
total_adultos = len(adultos)
total_adultos_supervivientes = len(AdultosSuperviventes)
total_no_supervivientes = total_adultos - total_adultos_supervivientes 

#Datos para la grafica
etiquetas =  ['Adultos Supervivientes', 'Adultos No Supervivientes']
valores = [total_adultos_supervivientes,total_no_supervivientes]
colores = ['#4CAF50','#E57373']

#Creagar Grafica 
plt.figure(figsize=(6,4))
plt.pie(
    valores,
    labels=etiquetas,
    colors=colores,
    shadow=True,
    autopct='%1.1f%%',
    startangle=90,
    explode=(0.05,0)
    )
plt.title("Compracion: Adultos Totales vs Adultos Supervivientes")
plt.show()
```

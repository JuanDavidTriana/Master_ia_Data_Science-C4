```python
import pandas as pd

df = pd.read_csv("house_prices.csv")

df.head()
```

```python
df.info()
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
pd.set_option('display.max_rows', 10)   
pd.set_option('display.max_columns', 10)

print(df["Society"].value_counts())
```

```python
df["Society"] = df["Society"].fillna("SinSociedad")
```

```python
df = df.drop(columns=["Plot Area","Dimensions"])
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
pd.set_option('display.max_rows', 5)   
pd.set_option('display.max_columns', 5)
print(df["Car Parking"].value_counts())
```

```python
df[["Parking_Count", "Parking_Type"]] = df["Car Parking"].str.split(pat=" ", n=1, expand=True)
df["Parking_Type"] = df["Parking_Type"].str.replace(",","", regex=False)
```

```python
print(df["Parking_Type"].value_counts())
```

```python
df["Parking_Count"].value_counts()
```

```python
import numpy as np
df["Parking_Count"] = pd.to_numeric(df["Parking_Count"])
df.loc[df["Parking_Count"] > 6, "Parking_Count"] = np.nan
```

```python
df["Parking_Count"].value_counts()
```

```python
df["Parking_Count"] = df["Parking_Count"].fillna(0)
```

```python
df["Parking_Count"].value_counts()
```

```python
df["Parking_Type"].value_counts()
df["Parking_Type"] = df["Parking_Type"].fillna("SinParquedero")
```

```python
df["Parking_Type"].value_counts()
```
